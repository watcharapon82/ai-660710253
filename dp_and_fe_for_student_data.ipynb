```python
data_clean_up = data.copy()
```

```python
data_clean_up.dtypes
```

```python
def gender_std(x):

  if x in ["Male", "ช", "M"]:
    return "ชาย"
  elif x in ["Female", "ญ", "F", "Female"]:
    return "หญิง"
  else:
    return x
```

```python
def faculty_std(x):
  if x in ["วิทย์", "คณะวิทยาศาสตร์", "Sci", "Science"]:
    return "วิทยาศาสตร์"
  elif x in ["วิศวฯ", "วิศวกรรม", "วิศวะ", "Engineering"]:
    return "วิศวกรรมศาสตร์"
  elif x in ["ศิลป์", "Arts", "Liberal Arts"]:
    return "ศิลปศาสตร์"
  else:
    return x
```

```python
Q1 = data_clean_up['อายุ'].quantile(0.25)
Q3 = data_clean_up['อายุ'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
def age_std(x):
  if x <lower_bound or x > upper_bound:
    return data_clean_up['อายุ'].mean()
  else:
    return x
```

```python
data_clean_up['เพศ'] = data_clean_up['เพศ'].apply(gender_std)
```

```python
data_clean_up['คณะ'] = data_clean_up['คณะ'].apply(faculty_std)
```

```python
data_clean_up['อายุ'] = data_clean_up['อายุ'].apply(age_std)
```

```python
data_clean_up["GPA"] = data_clean_up["GPA"].abs()
```

```python
n1 = data_clean_up["อายุ"].mean()
n2 = data_clean_up["GPA"].mean()
n3 = data_clean_up["ส่วนสูง_cm"].mean()
n4 = data_clean_up["น้ำหนัก_kg"].mean()
n5 = data_clean_up["ชั่วโมงเรียน_ต่อสัปดาห์"].mean()
```

```python
data_clean_up["อายุ"] = data_clean_up["อายุ"].fillna(n1)
data_clean_up["GPA"] =data_clean_up["GPA"].fillna(n2)
data_clean_up["ส่วนสูง_cm"] =data_clean_up["ส่วนสูง_cm"].fillna(n3)
data_clean_up["น้ำหนัก_kg"] =data_clean_up["น้ำหนัก_kg"].fillna(n4)
data_clean_up["ชั่วโมงเรียน_ต่อสัปดาห์"] =data_clean_up["ชั่วโมงเรียน_ต่อสัปดาห์"].fillna(n5)
```

```python
data_clean_up["เพศ"] = data_clean_up["เพศ"].fillna(data_clean_up["เพศ"].mode()[0])
data_clean_up["คณะ"] = data_clean_up["คณะ"].fillna(data_clean_up["คณะ"].mode()[0])
```

```python
data_clean_up = data_clean_up.dropna()
```

```python
student_health = data_clean_up.copy()
student_health['BMI'] = student_health['น้ำหนัก_kg'] / (student_health['ส่วนสูง_cm']/100)**2
student_health['สุขภาพ'] = pd.cut(student_health['BMI'],
                                   bins=[0, 18.5, 25, 30, 100],
                                   labels=['ผอม', 'ปกติ', 'อ้วน', 'อ้วนมาก'])

student_health
```
